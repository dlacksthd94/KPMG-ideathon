```python
import pandas as pd 
import numpy as np 
```

```python
path_이차전지 = '/home/kic/data/kodex/KODEX_2차전지산업_20230207.csv'
path_미래차 = '/home/kic/data/kodex/KODEX_K-미래차액티브_20230207.csv'
path_신재생 = '/home/kic/data/kodex/KODEX_K-신재생에너지액티브_20230207.csv'
path_dart = '/home/kic/data/rpt_list.csv'
```

```python
dart_dtype = {'corp_code':'object',
              'corp_name': 'object', 
              'stock_code': 'object',
              'corp_cls': 'object',
              'report_nm': 'object',
              'rcept_no': 'object',
              'flr_nm': 'object',
              'rcept_dt': 'object',
              'rm': 'object'}
```

```python
이차전지 = pd.read_csv(path_이차전지)
미래차 = pd.read_csv(path_미래차)
신재생 = pd.read_csv(path_신재생)
dart = pd.read_csv(path_dart, dtype=dart_dtype)
```

```python
kodex_concat = pd.concat([이차전지, 미래차, 신재생])
print(len(kodex_concat), len(kodex_concat.drop_duplicates(subset='종목코드')))
```

```python
kodex_concat = kodex_concat.drop_duplicates(subset='종목코드')
```

```python
kodex_merged = pd.merge(kodex_concat, dart, left_on='종목코드', right_on='stock_code', how='right')
kodex_merged = kodex_merged.dropna().reset_index()
kodex_merged.drop(['번호', 'index', '종목명', '종목코드'], axis = 1, inplace=True)
kodex_merged.head()
```

```python
kodex_merged.sort_values(by=['flr_nm', 'rcept_dt'], inplace=True, ascending=False)
kodex_merged = kodex_merged.drop_duplicates(subset=['flr_nm'], keep='first')
```
