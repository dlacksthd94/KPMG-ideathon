```python
import pandas as pd 
import numpy as np 
```

```python
path_이차전지 = '/home/kic/data/kodex/KODEX_2차전지산업_20230207.csv'
path_탄소효율 = '/home/kic/data/kodex/KODEX_탄소효율그린뉴딜_20230207.csv'
path_dart = '/home/kic/data/rpt_list.csv'
```

```python
dart_dtype = {'corp_code':'object',
              'corp_name': 'object', 
              'stock_code': 'object',
              'corp_cls': 'object',
              'report_nm': 'object',
              'rcept_no': 'object',
              'flr_nm': 'object',
              'rcept_dt': 'object',
              'rm': 'object'}
```

```python
이차전지 = pd.read_csv(path_이차전지)
탄소효율 = pd.read_csv(path_탄소효율)
dart = pd.read_csv(path_dart, dtype=dart_dtype)
```

```python
이차전지.head()
```

```python
dart.head()
```

```python
이차전지_merged = pd.merge(이차전지, dart, left_on='종목코드', right_on='stock_code', how='right')
이차전지_merged = 이차전지_merged.dropna().reset_index()
이차전지_merged.drop(['번호', 'index'], axis = 1, inplace=True)
이차전지_merged.head()
```

```python
탄소효율_merged = pd.merge(탄소효율, dart, left_on='종목코드', right_on='stock_code', how='right')
탄소효율_merged = 탄소효율_merged.dropna().reset_index()
탄소효율_merged.drop(['번호', 'index'], axis = 1, inplace=True)
탄소효율_merged
```
