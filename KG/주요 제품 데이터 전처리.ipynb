```python
import re

def processing_word(word):
    r1 = re.compile(r'\b\S+\b(?:\s\b\S+\b){1,6}')
    r2 = re.compile(r'(?!\S{2})\S+')
    stop_words = ['기타', '-', '상품']

#     processed_lst = []
#     for i in range(len(word_lst)):
#     if re.sub(r'\s+', '', word):
#         return

    if '기타' in word:
        return

    if word in stop_words :
        return

    elif r1.match(word) and r2.match(word):
        word = re.sub(r'\s+', '', word)
#         processed_lst.append(word_lst[i])

    elif ',' in word:
        temp = word.split(',')
        temp = [x.strip() for x in temp]
        word=temp
#         processed_lst.extend(temp)

    elif '(' in word:
        word = word.split('(')[0].strip()
#         processed_lst.append(word_lst[i])

    elif ' 외' in word:
        word = re.sub(' 외', '', word)
    else:
#         processed_lst.append(word_lst[i])
        pass
#     print(word)
        
    return word

```

```python
# corp_code와 합치기 
주요제품_level1_processed = dict()
for key, value in data1.items():
    products = list(value.keys())
    temp = [processing_word(product) for product in products]
    temp = [word for word in temp if word is not None and (len(word)>1)]
    # flatten list of lists
    temp2 = []
    for obj in temp:
        if type(obj) == list:
            temp2 += obj
        else:
            temp2.append(obj)
    주요제품_level1_processed[key] = temp2
```
